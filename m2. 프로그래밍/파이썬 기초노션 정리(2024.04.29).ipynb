    - **1. 기본 데이터 타입**
        
        데이터는 다양한 형태로 존재하며, 각 형태는 **데이터 타입**이라고 불립니다. 데이터 타입은 데이터의 특성과 용도를 정의하며, 프로그램에서 데이터를 효율적으로 처리하기 위해 중요합니다.
        
        - **숫자형:**
            - **정수형 (int):** 소수점 없는 숫자 (예: 1, 2, 3)
            - **실수형 (float):** 소수점 포함 숫자 (예: 1.2, 3.14)
            - **복소수형 (complex):** 실수부와 허수부로 구성된 숫자 (예: 1+2j)
        - **문자열형 (str):** 따옴표( ’’,”” ) 안에 포함된 문자열 (예: "Hello", 'World')
        - **논리형 (bool):** True 또는 False 값을 가짐(대소문자 구분함)
        - **None:값이 없는 것을 나타냄 (예: 변수가 초기화되지 않은 경우)**

    - **3.f-문자열과 함께 사용하여 더욱 간결하게(문자열 포매팅)**
        
        파이썬에서 문자열 포매팅은 다양한 방식으로 가능합니다. 각 방법은 장단점이 존재하며, 상황에 따라 적절한 방법을 선택하는 것이 중요합니다.
        
        - **1. `%` 연산자**
            
            C 스타일 포매팅 방식으로, 오랫동안 사용되어 왔습니다.
            
            ### **장점**
            
            - 간단하고 직관적
            - 다양한 형식 지정 옵션 지원
            
            ### **단점**
            
            - 복잡한 형식 지정 시 가독성이 떨어짐
            - 최신 버전에서 권장되지 않음
            
            ### **예시**
            
            **Python**
            
            ```jsx
            name = "홍길동"
            age = 30
            
            # 이름과 나이를 문자열에 결합
            message = "%s님은 %d살입니다." % (name, age)
            print(message)  # 출력: 홍길동님은 30살입니다.
            
            # 정렬 및 자릿수 지정
            message = "%-10s님은 %3d살입니다." % (name, age)
            print(message)  # 출력: 홍길동님은  30살입니다.
            
            # 소수점 표기
            price = 1234.56789
            
            message = "상품 가격은 %.2f입니다." % price
            print(message)  # 출력: 상품 가격은 1234.57입니다.
            ```
            
        - **2. `format()` 메서드**
            
            `%` 연산자보다 더 새롭고 명확하며, 다양한 기능을 제공합니다.
            
            ### **장점**
            
            - 명확하고 가독성이 높음
            - 다양한 객체 포맷팅 지원
            - 포맷 문자열을 따로 정의할 수 있음
            
            ### **단점**
            
            - `%` 연산자보다 다소 복잡
            - f-string보다 간결하지 않을 수 있음
            
            ### **예시**
            
            **Python**
            
            ```jsx
            name = "홍길동"
            age = 30
            
            # 이름과 나이를 문자열에 결합
            message = "{}님은 {}살입니다.".format(name, age)
            print(message)  # 출력: 홍길동님은 30살입니다.
            
            # 정렬 및 자릿수 지정
            message = "{:<10}님은 {:3}살입니다.".format(name, age)
            print(message)  # 출력: 홍길동님은  30살입니다.
            
            # 소수점 표기
            price = 1234.56789
            
            message = "상품 가격은 {:.2f}입니다.".format(price)
            print(message)  # 출력: 상품 가격은 1234.57입니다.
            ```

            `name = "홍길동"

            # 이름과 나이를 문자열에 결합

- **03)파이썬 기본 집합형 자료형**
    - **1. 리스트 (List)**
        - 순서 있는 값들의 모음
        - 중복 허용
        - 변경 가능
        
        **예시:**
        
        **Python**
        
        ```jsx
        # 리스트 생성
        my_list = [1, 2, 3, "apple", "banana"]
        
        # 리스트 길이
        len(my_list)
        
        # 특정 인덱스 값 접근
        my_list[2]
        
        # 리스트 값 추가
        my_list.append("orange")
        
        # 리스트 값 삭제
        my_list.remove("banana")
        
        # 리스트 순서 뒤집기
        my_list.reverse()
        ```
        
    - **2. 튜플 (Tuple)**
        - 순서 있는 값들의 모음
        - 중복 허용
        - 변경 불가능
        
        **예시:**
        
        **Python**
        
        ```jsx
        # 튜플 생성
        my_tuple = (1, 2, 3, "apple", "banana")
        
        # 튜플 길이
        len(my_tuple)
        
        # 특정 인덱스 값 접근
        my_tuple[2]
        
        # 튜플 값 변경 불가능
        my_tuple[0] = "orange" # TypeError 발생
        
        # 튜플 언패킹
        a, b, c, d, e = my_tuple
        ```
        
    - **3. 딕셔너리 (Dictionary)**
        - 키-값 쌍으로 구성된 자료형

        - 변경 가능

- **05)파이썬 if문: 조건에 따라 코드를 실행하는 분기 구조**
    
    파이썬 if문은 특정 조건이 만족될 때만 코드를 실행할 수 있도록 하는 **분기 구조**입니다. 마치 게임에서 특정 레벨을 클리어해야 다음 스테이지로 진행할 수 있는 것과 같죠.
    
    - **1. 기본 구조:**
        
        **Python**
        
        ```jsx
        if 조건식:
            # 조건이 참일 때 실행할 코드
        ```
        
        - **조건식:** 논리 표현으로, 참(True) 또는 거짓(False)으로 평가됩니다.
        - **코드 블록:** 들여쓰기를 통해 구분되는, 조건이 참일 때 실행되는 코드들
    - **2. 비교 연산자:**
        
        
        | 연산자 | 의미 | 예시 |
        | --- | --- | --- |
        | == | 같다 | 10 == 10 (참) |
        | != | 같지 않다 | 10 != 10 (거짓) |
        | > | 크다 | 10 > 5 (참) |
        | < | 작다 | 5 < 10 (참) |
        | >= | 크거나 같다 | 10 >= 10 (참) |
        | <= | 작거나 같다 | 5 <= 10 (참) |
        |  |  |  |
    - **3. 논리 연산자:**
        
        
        | 연산자 | 의미 | 예시 |
        | --- | --- | --- |
        | and | 두 조건 모두 만족해야 함 | (age >= 18) and (gender == "male") |
        | or | 두 조건 중 하나라도 만족하면 됨 | (age >= 18) or (gender == "female") |
        | not | 조건의 반전 | not (age >= 18) |
    - **4.if~else 문**
        - **기본 구조:**
            
            **Python**
            
            ```jsx
            if 조건:
                # 조건이 참일 때 실행할 코드
            else:
                # 조건이 거짓일 때 실행할 코드
            ```
            
            **예시:**
            
            **Python**
            
            ```jsx
            # 숫자가 양수인지 음수인지 판별하는 코드
            num = 10
            
            if num > 0:
                print("양수입니다.")
            else:
                print("음수입니다.")
            ```
            
        - **else if:**
            
            **Python**
            
            ```jsx
            # 숫자가 0, 양수, 음수인지 판별하는 코드
            num = 0
            
            if num == 0:
                print("0입니다.")
            elif num > 0:
                print("양수입니다.")
            else:
                print("음수입니다.")
            ```
            
        - **중첩된 if:**
            
            **Python**
            
            ```jsx
            # 숫자가 짝수인지 홀수인지, 양수인지 음수인지 판별하는 코드
            num = 11
            
            if num % 2 == 0:
                print("짝수입니다.")
            else:
                print("홀수입니다.")
            
            if num > 0:
                print("양수입니다.")
            else:
                print("음수입니다.")
            ```
            
        
    - **5.if~elif~else 문**
        
        **if~elif~else** 문은 여러 개의 조건을 차례로 검사하여 해당 조건에 맞는 코드 블록만 실행하는 데 사용됩니다.  **if** 문만 사용하는 것보다 복합적인 상황을 처리할 때 유용합니다.
        
        **기본 구조:**
        
        ```jsx
        if 조건 1:
          # 조건 1이 참일 때 실행할 코드
        elif 조건 2:
          # 조건 2가 참일 때 실행할 코드 (조건 1은 거짓)
        else:
          # 모든 조건이 거짓일 때 실행할 코드
        ```
        
        **예시:**
        
        **Python**
        
        ```jsx
        # 학점을 판별하는 코드
        grade = 85
        
        if grade >= 90:
          print("A 학점")
        elif grade >= 80:
          print("B 학점")
        else:
          print("C 학점 이하")
        ```
        
        **작동 방식:**
        
        1. 프로그램은 처음에 `if` 조건을 평가합니다.
        2. 만약 `if` 조건이 참이면, 해당 조건 블록의 코드만 실행되고 나머지 부분은 스킵됩니다.
        3. `if` 조건이 거짓이면, 다음 `elif` 조건으로 넘어갑니다.
        4. 만약 해당하는 `elif` 조건이 참이면, 해당 조건 블록의 코드만 실행되고 나머지 부분은 스킵됩니다.
        5. 모든 `if` 와 `elif` 조건이 거짓이면, 마지막 `else` 블록의 코드가 실행됩니다.
        
        **중요 사항:**
        
        - `elif` 는 **하나의 `if` 문** 에만 연결될 수 있습니다. 여러 `if` 문에 `elif` 를 사용하고 싶다면, 중첩된 `if` 구조를 이용해야 합니다.
        - `elif` 는 **여러 개** 사용할 수 있습니다.
        - 조건의 순서는 중요합니다. 먼저 평가된 조건에 따라 해당 코드 블록만 실행되기 때문입니다.
        - 모든 조건에 해당하는 코드가 없는 경우, 마지막 `else` 블록을 사용할 수 있습니다.
    - **6.주의 사항:**
        - 조건은 `True` 또는 `False` 값을 반환하는 논리 표현식이어야 합니다.
        - 들여쓰기를 사용하여 코드 블록을 명확하게 구분해야 합니다.
        - 조건의 순서는 중요합니다. 먼저 평가되는 조건부터 코드가 실행됩니다.
        - 모든 조건에 해당하는 코드가 없는 경우, 마지막 `else` 블록을 사용할 수 있습니다.
- **06)파이썬 for문**
    
    for 문은 컴퓨터 프로그래밍에서 반복적인 작업을 간편하게 수행할 수 있도록 하는 강력한 도구입니다. 특정 조건이 만족되는 동안 코드 블록을 반복적으로 실행하여 시간을 절약하고 코드의 효율성을 높일 수 있습니다.
    
    - **for 문의 구성 요소:**
        - **초기화**: 반복 시작 전에 한 번만 실행되는 코드입니다. 일반적으로 루프 변수를 선언하고 초기값을 할당하는 데 사용됩니다.
        - **조건**: 반복을 지속할지 여부를 결정하는 논리적 표현입니다. 조건이 참(True)인 경우 코드 블록을 실행하고, 거짓(False)인 경우 루프를 종료합니다.
        - **증감**: 반복이 끝날 때마다 실행되는 코드입니다. 일반적으로 루프 변수의 값을 증가하거나 감소시키는 데 사용됩니다.
        - **코드 블록**: 조건이 참인 동안 반복적으로 실행되는 코드입니다. 여러 줄의 코드를 포함할 수 있으며, 들여쓰기를 통해 구분됩니다.
    - **for 문의 기본 형식:**
        
        **Python**
        
        ```jsx
        for <루프 변수> in <반복 범위>:
            <코드 블록>
        ```
        
    - **for 문 활용 예시:**
        - **1부터 10까지 숫자 출력하기:**
            
            **Python**
            
            ```jsx
            for i in range(1, 11):
                print(i)
            ```
            
        - **배열의 모든 원소 출력하기:**
            
            **Python**
            
            ```jsx
            arr = [1, 2, 3, 4, 5]
            for element in arr:
                print(element)
            ```
            
        - **짝수만 출력하기:**
            
            **Python**
            
            ```jsx
            for i in range(2, 11, 2):
                print(i)
            ```
            
        - **for 문으로 리스트를 반복 처리하기:**
            
            **Python**
            
            ```jsx
            for i in 리스트변수:
                # 리스트의 각 항목에 대한 작업을 수행합니다.
            ```
            
            **예시:**
            
            **Python**
            
            ```jsx
            # 리스트의 모든 항목을 출력합니다.
            my_list = ["사과", "바나나", "오렌지"]
            for fruit in my_list:
                print(fruit)
            ```
            
        - **for 문을 사용하여 딕셔너리를 반복 처리하기:**
            
            **Python**
            
            ```jsx
            for key, value in 딕셔너리변수.items():
                # 딕셔너리의 각 키와 값에 대한 작업을 수행합니다.
            ```
            
            **예시:**
            
            **Python**
            
            ```jsx
            # 딕셔너리의 모든 키와 값을 출력합니다.
            my_dict = {"name": "John Doe", "age": 30}
            ```

        - **for 문을 사용하여 문자열을 반복 처리하기:**
            
            **Python**
            
            `for 문자 in 문자열`
            
    - **for 문의 고급 기능:**
        - **continue**: 현재 반복만 건너뛰고 다음 반복으로 이동합니다.
        - **break**: 루프를 완전히 종료합니다.
        - **else**: 루프 조건이 False가 되면 실행되는 코드 블록입니다.
    - **for 문 활용 팁:**
        - for 문은 반복적인 작업을 간편하게 수행할 수 있도록 하는 강력한 도구입니다.
        - for 문을 사용하면 코드의 효율성을 높이고 가독성을 향상시킬 수 있습니다.
        - for 문의 다양한 기능을 활용하여 다양한 반복 작업을 수행할 수 있습니다.
    - **for문 관련 함수**
        - **enumerate(): 리스트의 인덱스와 값 접근**
            
            enumerate() 함수는 리스트의 각 항목에 대한 인덱스와 값을 튜플로 반환합니다.
            
            **예시:**
            
            **Python**
            
            ```jsx
            my_list = ["사과", "바나나", "오렌지"]
            for index, fruit in enumerate(my_list):
                print(f"인덱스: {index}, 값: {fruit}")
            ```
            
            **출력:**
            
            ```jsx
            인덱스: 0, 값: 사과
            인덱스: 1, 값: 바나나
            인덱스: 2, 값: 오렌지
            ```
            
        - **zip(): 여러 반복 가능한 객체를 하나로 결합**
            
            zip() 함수는 여러 반복 가능한 객체를 하나의 반복 가능한 객체로 결합합니다. 각 반복에서 각 객체의 첫 번째 항목, 두 번째 항목 등을 묶어 튜플을 만듭니다.
            
            **예시:**
            
            **Python**
            
            ```jsx
            names = ["John", "Jane", "Jack"]
            ages = [30, 25, 40]
            for name, age in zip(names, ages):
                print(f"{name}은 {age}살입니다.")
            ```
            
            **출력:**
            
            ```jsx
            John은 30살입니다.
            Jane은 25살입니다.
            Jack은 40살입니다.
            ```
            
        - **reversed(): 리스트를 반대로 반복**
            
            reversed() 함수는 리스트를 반대로 반복할 수 있도록 합니다.
            
            **예시:**
            
            **Python**
            
            ```jsx
            my_list = [1, 2, 3, 4, 5]
            for num in reversed(my_list):
                print(num)
            ```
            
            **출력:**
            
            ```jsx
            5
            4
            3
            2
            1
            ```
            
        - **sorted(): 리스트를 정렬**
            
            sorted() 함수는 리스트를 정렬합니다. 기본적으로 오름차순으로 정렬하지만, reverse=True 옵션을 사용하면 내림차순으로 정렬할 수도 있습니다.
            
            **예시:**
            
            **Python**
            
            ```jsx
            my_list = [5, 2, 4, 1, 3]
            for num in sorted(my_list):
                print(num)
            ```
            
            **출력:**
            
            ```jsx
            1
            2
            3
            4
            5
            ```
            
        - **filter(): 조건에 맞는 항목만 선택**
            
            filter() 함수는 조건에 맞는 항목만 선택하여 새로운 리스트를 만듭니다.
            
            **예시:**
            
            **Python**
            
            ```jsx
            my_list = [1, 2, 3, 4, 5]
            even_numbers = list(filter(lambda x: x % 2 == 0, my_list))
            print(even_numbers)
            ```
            
            **출력:**
            
            ```jsx
            [2, 4]
            ```
            
        - **map(): 각 항목에 함수 적용**
            
            map() 함수는 각 항목에 함수를 적용하여 새로운 리스트를 만듭니다.
            
            **예시:**
            
            **Python**
            
            ```jsx
            my_list = [1, 2, 3, 4, 5]
            squared_numbers = list(map(lambda x: x ** 2, my_list))
            print(squared_numbers)
            ```
            
            **출력:**
            
            ```jsx
            [1, 4, 9, 16, 25]
            ```
            
        - **itertools 모듈**
            - product(): 여러 반복 가능한 객체의 모든 조합을 생성합니다.
            - permutations(): 여러 반복 가능한 객체의 모든 순열을 생성합니다.
            - combinations(): 여러 반복 가능한 객체의 모든 조합을 생성합니다.
- **07)파이썬 while문**
    
    while 문은 조건이 참인 동안 반복적으로 코드 블록을 실행하는 제어문입니다. 조건이 거짓이 되면 반복을 멈춥니다.
    
    - **while 문의 기본 구조:**
        
        **Python**
        
        ```jsx
        while <조건>:
            <코드 블록>
        ```
        
    - **while 문의 작동 방식:**
        1. 조건을 확인합니다.
        2. 조건이 참이면 코드 블록을 실행합니다.
        3. 코드 블록을 실행한 후 다시 조건을 확인합니다.
        4. 조건이 참이면 2번으로 돌아갑니다.
        5. 조건이 거짓이면 반복을 멈춥니다.
    - **while 문 예시:**
        
        **Python**
        
        ```jsx
        # 1부터 10까지 숫자 출력
        i = 1
        while i <= 10:
            print(i)
            i += 1
        
        # 사용자 입력값을 받아 계속 더하는 프로그램
        total = 0
        while True:
            value = input("숫자를 입력하세요 (종료하려면 q를 입력): ")
            if value == "q":
                break
            total += int(value)
        print(f"총합: {total}")
        ```
        
    - **while 문 활용 팁:**
        - 무한 루프를 만들 때 유용합니다.
        - 사용자 입력을 반복적으로 받아야 할 때 사용합니다.
        - 특정 조건이 충족될 때까지 반복해야 할 때 사용합니다.
    - **while 문 주의 사항:**
        - 무한 루프에 빠지지 않도록 조건을 잘 설정해야 합니다.
        - 조건 검사에 시간이 오래 걸리는 경우 성능 저하를 초래할 수 있습니다.
